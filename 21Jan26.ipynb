```
import pandas as pd
import os
import numpy as np

# 1. Corrected Path
data_folder = "/content/drive/Othercomputers/My Computer/Documents/Insulin/actigraphy/VectorMagnitudeData/2.4"

summary_rows = []

if not os.path.exists(data_folder):
    print(f"‚ùå Folder not found! Please check the mount. Path tried: {data_folder}")
else:
    files = [f for f in os.listdir(data_folder) if f.endswith((".xlsx", ".xls"))]
    print(f"Found {len(files)} files. Starting processing...")

    for file in files:
        filepath = os.path.join(data_folder, file)
        subject_id = os.path.splitext(file)[0]

        xls = pd.ExcelFile(filepath)

        for sheet in xls.sheet_names:
            # Read sheet without assuming a header row
            df_raw = pd.read_excel(xls, sheet_name=sheet, header=None)

            # --- DATA EXTRACTION ---
            # Treatment: Cell C1 -> Row 0, Column 2
            treatment = df_raw.iloc[0, 2]

            # VM Data: Column C starting from Row 4 -> Column index 2, Row index 3 onwards
            vm = pd.to_numeric(df_raw.iloc[3:, 2], errors='coerce').dropna()

            if vm.empty:
                continue

            # --- CALCULATIONS ---
            total_vm = vm.sum()
            mean_vm = vm.mean()
            prop_zero = (vm == 0).mean()

            # Mean activity during active minutes ONLY (ignore the zeros)
            active_only = vm[vm > 0]
            mean_active_vm = active_only.mean() if not active_only.empty else 0

            # --- INACTIVITY BOUT ANALYSIS ---
            # Identify sequences of zeros
            is_zero = (vm == 0).astype(int)
            # Create a "group" ID that changes whenever the value changes (0 to >0 or vice-versa)
            abs_diff = is_zero.diff().ne(0).cumsum()
            # Filter for groups that are zeros and count their lengths
            zero_bouts = is_zero[is_zero == 1].groupby(abs_diff).count()

            max_inactive_period = zero_bouts.max() if not zero_bouts.empty else 0
            mean_inactive_length = zero_bouts.mean() if not zero_bouts.empty else 0
            num_inactive_bouts = len(zero_bouts)

            # --- SAVE DATA ---
            summary_rows.append({
                "SubjectID": subject_id,
                "Treatment": treatment,
                "TimePoint": sheet,
                "TotalVM": total_vm,
                "MeanVM": mean_vm,
                "MeanActiveVM": mean_active_vm,
                "PropZero": prop_zero,
                "MaxInactiveBout": max_inactive_period,
                "MeanInactiveBout": mean_inactive_length,
                "NumInactiveBouts": num_inactive_bouts
            })

    # 4. Save the Master File
    summary_df = pd.DataFrame(summary_rows)
    output_path = os.path.join(data_folder, "MasterSummary_Actigraphy.csv")
    summary_df.to_csv(output_path, index=False)
    print(f"\n‚úÖ Pipeline Complete! File saved to: {output_path}")
```

```
import pandas as pd
import os
import numpy as np

# 1. Path setup
data_folder = "/content/drive/Othercomputers/My Computer/Documents/Insulin/actigraphy/VectorMagnitudeData/2.4"

summary_rows = []

if not os.path.exists(data_folder):
    print(f"‚ùå Folder not found! Path tried: {data_folder}")
else:
    # Filter out hidden files (those starting with '.') and keep only excel
    files = [f for f in os.listdir(data_folder) if f.endswith((".xlsx", ".xls")) and not f.startswith('.')]
    print(f"Found {len(files)} valid Excel files. Starting processing...")

    for file in files:
        filepath = os.path.join(data_folder, file)
        subject_id = os.path.splitext(file)[0]

        try:
            # Explicitly using openpyxl engine for .xlsx files
            xls = pd.ExcelFile(filepath, engine='openpyxl')

            for sheet in xls.sheet_names:
                df_raw = pd.read_excel(xls, sheet_name=sheet, header=None)

                # --- DATA EXTRACTION ---
                treatment = df_raw.iloc[0, 2] # Cell C1

                # VM Data (Column C, Row 4+)
                vm = pd.to_numeric(df_raw.iloc[3:, 2], errors='coerce').dropna()

                # Time Data (Column B, Row 4+)
                # We convert to 'datetime' then extract just the 'time'
                times = pd.to_datetime(df_raw.iloc[3:, 1], errors='coerce').dt.time

                if vm.empty or times.dropna().empty:
                    continue

                # --- CIRCADIAN SPLIT (6am - 6pm) ---
                day_start = pd.to_datetime('06:00:00').time()
                day_end = pd.to_datetime('18:00:00').time()

                is_day = [(t >= day_start and t < day_end) if pd.notnull(t) else False for t in times]
                is_night = [not val for val in is_day]

                vm_day = vm[is_day]
                vm_night = vm[is_night]

                # --- BASIC STATS ---
                total_vm = vm.sum()
                mean_vm = vm.mean() # Total mean (including zeros)
                prop_zero = (vm == 0).mean()

                # --- INACTIVITY BOUT ANALYSIS ---
                is_zero = (vm == 0).astype(int)
                abs_diff = is_zero.diff().ne(0).cumsum()
                zero_bouts = is_zero[is_zero == 1].groupby(abs_diff).count()

                max_inactive_period = zero_bouts.max() if not zero_bouts.empty else 0
                num_inactive_bouts = len(zero_bouts)

                # --- SAVE DATA ---
                summary_rows.append({
                    "SubjectID": subject_id,
                    "Treatment": treatment,
                    "TimePoint": sheet,
                    "TotalVM": total_vm,
                    "MeanVM_Total": mean_vm,
                    "Mean_Day": vm_day.mean() if not vm_day.empty else 0,
                    "Mean_Night": vm_night.mean() if not vm_night.empty else 0,
                    "MaxInactiveBout": max_inactive_period,
                    "NumInactiveBouts": num_inactive_bouts,
                    "PropZero": prop_zero
                })
        except Exception as e:
            print(f"‚ö†Ô∏è Could not read file {file}: {e}")

    # Create Master File
    if summary_rows:
        summary_df = pd.DataFrame(summary_rows)
        output_path = os.path.join(data_folder, "MasterSummary_Final.csv")
        summary_df.to_csv(output_path, index=False)
        print(f"\n‚úÖ Success! File saved to: {output_path}")
    else:
        print("No data was processed. Check if the sheets match the expected format.")
```

```
import pandas as pd
import os

# 1. Load the summary
base_path = "/content/drive/Othercomputers/My Computer/Documents/Insulin/actigraphy/VectorMagnitudeData"
df = pd.read_csv(os.path.join(base_path, "Grand_Master_Summary.csv"))

# 2. Create the Pivot Table naturally (it will be alphabetical for now)
pivot_df = df.pivot_table(index='TimePoint',
                          columns='Treatment',
                          values='MeanVM_Total',
                          aggfunc='mean')

# 3. List the order you want.
# NOTE: Make sure these match your Excel cells EXACTLY (case sensitive)
my_order = ['BL', '5dpi', '12dpi', '19dpi', '26dpi']

# 4. Reindex: This just moves rows around without deleting data
# We only reindex with labels that actually exist in the pivot table index
existing_labels = [label for label in my_order if label in pivot_df.index]
other_labels = [label for label in pivot_df.index if label not in my_order]

ordered_pivot = pivot_df.reindex(existing_labels + other_labels)

# 5. SAVE TO EXCEL
output_file = os.path.join(base_path, "Ordered_Pivot_Summary.xlsx")
ordered_pivot.to_excel(output_file)

print("--- Ordered Pivot Table ---")
display(ordered_pivot)

print("\n--- Diagnostic: Every unique label found in your CSV ---")
print(df['TimePoint'].unique())
```
